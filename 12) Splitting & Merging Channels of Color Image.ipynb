```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "C:/Users/User/OpenCv/misc/"
imgpath1 = path + "lena_color_512.tif"
img = cv2.imread(imgpath1, 1)
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
r, g, b = cv2.split(img1)
titles = ["original image", "Red", "Green", "Blue"]
images = [cv2.merge((r, g, b)), r, g, b]
mapping = ["", "Reds", "Greens", "Blues"]
```

```python
for i in range(len(images)):
    plt.subplot(1, 4, i + 1)
    if(i == 0):
        plt.imshow(images[i])
    else:
        plt.imshow(images[i], cmap = mapping[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
