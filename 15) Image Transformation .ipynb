## 4) Rotating an Image

```python
rows, columns, channels = img.shape
angle = 360
while(1):
    if(angle == 0):
        angle = 360
        
    R = cv2.getRotationMatrix2D((columns / 2, rows / 2), angle, 0.5)
    output = cv2.warpAffine(img, R, (columns, rows))
    cv2.imshow("Rotated Image", output)
    angle = angle - 1
    time.sleep(0.01)
    if(cv2.waitKey(1) == 27):
        break
cv2.destroyWindow("Rotated Image")
```

## 5) Rotation Effect on Live WebCam Feed

```python
windowName = "Live Video Feed"
cv2.namedWindow(windowName)
cap = cv2.VideoCapture(0)
if(cap.isOpened()):
    ret, frame = cap.read()
else:
    ret = False
    
rows, columns, channels = frame.shape
angle = 0
# scale = 1
scale = 0
while(1):
    ret, frame = cap.read()
    if(angle == 360):
        angle = 0
    if(scale < 2):
        scale = scale + 0.1
    if(scale >= 2):
        scale = 0.1
        
    R = cv2.getRotationMatrix2D((columns / 2, rows / 2), angle, scale)
    output = cv2.warpAffine(frame, R, (columns, rows))
    cv2.imshow(windowName, output)
    angle = angle + 1
    time.sleep(0.05)
    if(cv2.waitKey(1) == 27):
        break

cv2.destroyWindow(windowName)
cv2.destroyAllWindows()
cap.release()
```

## 6) Affine Transformation

```python
rows, columns, channels = img.shape
point1 = np.float32([[100, 100], [300, 100], [100, 300]])
point2 = np.float32([[200, 150], [400, 150], [200, 400]])
A = cv2.getAffineTransform(point1, point2)
output = cv2.warpAffine(img, A, (columns, rows))
plt.imshow(output)
plt.title("Rotated Image")
plt.show()
```

## 7) Perspective Transformation

```python
rows, columns, channels = img.shape
point1 = np.float32([[0, 0], [400, 0], [0, 400], [400, 400]])
point2 = np.float32([[0, 0], [500, 0], [0, 500], [500, 500]])
P = cv2.getPerspectiveTransform(point1, point2)
output = cv2.warpPerspective(img, P, (300, 300))
plt.imshow(output)
plt.title("Rotated Image")
plt.show()
```
