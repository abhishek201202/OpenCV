```python
import matplotlib.pyplot as plt
import cv2
```

```python
imgpath = "C:/Users/User/OpenCv/misc/4.2.07.tiff"
img = cv2.imread(imgpath, 0) ## reading in gray scale mode
```

```python
plt.imshow(img) ## default color maping
plt.title('Default color mapping')
plt.show()
## imshow doesn't actually know what type of 
## image is it taking into argument
## assign color by itself
```

```python
plt.imshow(img, cmap = 'Reds')
plt.title('Red color mapping')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
plt.imshow(img, cmap = 'gray')
plt.title('Gray color mapping')
plt.show()
```
