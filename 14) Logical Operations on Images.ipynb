```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = "C:/Users/User/OpenCv/misc/"
img1path = path + "4.2.06.tiff"
img2path = path + "lena_color_512.tif"

img1 = cv2.imread(img1path, 1)
img2 = cv2.imread(img2path, 1)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = cv2.bitwise_not(img1)
img4 = cv2.bitwise_and(img1, img2)
img5 = cv2.bitwise_or(img1, img2)
img6 = cv2.bitwise_xor(img1, img2)
```

```python
titles = ["img1", "img2", "not of img1", "and", "or", "xor"]
images = [img1, img2, img3, img4, img5, img6]
```

```python
for i in range(len(images)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
