```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "C:/Users/User/OpenCv/misc/"
imgpath = path + "lena_color_512.tif"

img = cv2.imread(imgpath, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## 1) Image Quantization 

```python
Z = img.reshape((-1, 3))
Z = np.float32(Z)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
```

```python
K = 2
ret, label1, center1 = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
center1 = np.uint8(center1)
res1 = center1[label1.flatten()]
output1 = res1.reshape((img.shape))
```

```python
K = 4
ret, label1, center1 = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
center1 = np.uint8(center1)
res1 = center1[label1.flatten()]
output2 = res1.reshape((img.shape))
```

```python
K = 12
ret, label1, center1 = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
center1 = np.uint8(center1)
res1 = center1[label1.flatten()]
output3 = res1.reshape((img.shape))
```

```python
output = [img, output1, output2, output3]
titles = ["Original Image", "k = 2", "k = 4", "k = 12"]
```

```python
for i in range(len(output)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

## 2) Kmeans Live WebCam

```python
w = 100
h = 100
cap = cv2.VideoCapture(0)
cap.set(3, w)
cap.set(4, h)
if(cap.isOpened()):
    ret, frame = cap.read()
else:
    ret = False
    
while(ret):
    ret, frame = cap.read()
    Z = frame.reshape((-1, 3))
    Z = np.float32(Z)
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    K = 12
    ret, label1, center1 = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
    center1 = np.uint8(center1)
    res1 = center1[label1.flatten()]
    output = res1.reshape((frame.shape))
    cv2.imshow("Original", frame)
    cv2.imshow("Quantize", output)
    if(cv2.waitKey(1) == 27):
        break
cv2.destroyAllWindows()
cap.release()
```
