```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
imgpath = "C:/Users/User/OpenCv/misc/cameraman.tif"
img = cv2.imread(imgpath, 0)
```

```python
th = 0
max_val = 255
ret, binary_inv = cv2.threshold(img, th, max_val, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

k = np.ones((5, 5), np.uint8)
erosion = cv2.erode(binary_inv, k, iterations = 3)
dilation = cv2.dilate(binary_inv, k, iterations = 1)
gradient = cv2.morphologyEx(binary_inv, cv2.MORPH_GRADIENT, k)

output = [img, binary_inv, erosion, dilation, gradient]
titles = ["Original", "Binary inv", "erosion", "dilations", "gradient"]

for i in range(len(output)):
    plt.subplot(3, 2, i + 1)
    plt.imshow(output[i], cmap = "gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
