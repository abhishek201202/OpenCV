```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "C:/Users/User/OpenCv/misc/"
img1path = path + "4.2.06.tiff"
img2path = path + "4.2.07.tiff"
img1 = cv2.imread(img1path, 1)
img2 = cv2.imread(img2path, 1)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
plt.title("Scenery")
plt.show()
```

```python
plt.imshow(img2)
plt.title("shimla_mirch")
plt.show()
```

```python
## plt.subplot(row, column, position of image >= 1)
plt.subplot(1, 2, 1)
plt.imshow(img1)

plt.subplot(1, 2, 2)
plt.imshow(img2)

plt.show()
```

```python
titles = ["Scenery", "shimla_mirch"]
images = [img1, img2]
for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
