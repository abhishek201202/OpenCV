```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "C:/Users/User/OpenCv/misc/"
imgpath = path + "4.2.07.tiff"
img = cv2.imread(imgpath)
```

## 1) Matplotlib to calculate and display the histogram

```python
plt.subplot(1, 2, 1)
plt.imshow(img, cmap = "gray")
plt.title("Image")
plt.xticks([])
plt.yticks([])


plt.subplot(1, 2, 2)
plt.hist(img.ravel(), 256, [0, 256])
# plt.hist(img.ravel(), no of beans, range of beans)
plt.title("Histogram")
plt.xlim(xmin = 0, xmax = 256)
plt.ylim(ymin = 0, ymax = 8000)
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
R, G, B = cv2.split(img)

plt.imshow(img, cmap = "gray")
plt.title("Image")
plt.xticks([])
plt.yticks([])
plt.show()

plt.subplot(3, 1, 1)
plt.hist(R.ravel(), 256, [0, 256])
plt.title("R Histogram")
plt.xlim(xmin = 0, xmax = 256)
plt.ylim(ymin = 0, ymax = 5000)
plt.show()

plt.subplot(3, 1, 2)
plt.hist(G.ravel(), 256, [0, 256])
plt.title("G Histogram")
plt.xlim(xmin = 0, xmax = 256)
plt.ylim(ymin = 0, ymax = 5000)
plt.show()

plt.subplot(3, 1, 3)
plt.hist(B.ravel(), 256, [0, 256])
plt.title("B Histogram")
plt.xlim(xmin = 0, xmax = 256)
plt.ylim(ymin = 0, ymax = 5000)
plt.show()
```

## 2) Numpy Histogram

```python
img = cv2.imread(imgpath, 1)
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.xticks([])
plt.yticks([])

plt.subplot(1, 2, 2)
hist, bins = np.histogram(img.ravel(), 256, [0, 256])
plt.xlim([0, 255])
plt.ylim([0, 8000])
plt.plot(hist)
plt.title("Histogram")
plt.show()
```

```python
img = cv2.imread(imgpath, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
R, G, B = cv2.split(img)


plt.imshow(img)
plt.title("Original Image")
plt.xticks([])
plt.yticks([])
plt.show()

plt.subplot(3, 1, 1)
hist, bins = np.histogram(R.ravel(), 256, [0, 256])
plt.xlim([0, 255])
plt.ylim([0, 8000])
plt.plot(hist, color= "r")
plt.title("R channel Histogram")
plt.show()

plt.subplot(3, 1, 2)
hist, bins = np.histogram(G.ravel(), 256, [0, 256])
plt.xlim([0, 255])
plt.ylim([0, 8000])
plt.plot(hist, color = "g")
plt.title("G channel Histogram")
plt.show()

plt.subplot(3, 1, 3)
hist, bins = np.histogram(B.ravel(), 256, [0, 256])
plt.xlim([0, 255])
plt.ylim([0, 8000])
plt.plot(hist, color = "b")
plt.title("B channel Histogram")
plt.show()
```

## 3) OpenCv Histogram

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
red_hist = cv2.calcHist([img], [0], None, [256], [0, 256])
green_hist = cv2.calcHist([img], [1], None, [256], [0, 256])
blue_hist = cv2.calcHist([img], [2], None, [256], [0, 256])
## [0], [1], [3] represent 0th, 1st and 2nd channels
```

## 4) Histogram Equalization

```python
## On GrayScale Image
imgpath = path + "lena_color_512.tif"
img = cv2.imread(imgpath, 0)

output1 = cv2.equalizeHist(img)

# clahe = cv2.createCLAHE()
clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize = (8, 8))
output2 = clahe.apply(img)


output = [img, output1, output2]
titles = ["Original image", "output1", "output2"]

for i in range(len(output)):
    plt.subplot(1, 3, i + 1)
    plt.imshow(output[i], cmap = "gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
## On Color Image
imgpath = path + "lena_color_512.tif"
img = cv2.imread(imgpath, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
R, G, B = cv2.split(img)

Routput = cv2.equalizeHist(R)
Goutput = cv2.equalizeHist(G)
Boutput = cv2.equalizeHist(B)
output1 = cv2.merge((Routput, Goutput, Boutput))

# clahe = cv2.createCLAHE()
clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize = (8, 8))
Rouptut = clahe.apply(R)
Gouptut = clahe.apply(G)
Bouptut = clahe.apply(B)
output2 = cv2.merge((Routput, Goutput, Boutput))

output = [img, output1, output2]
titles = ["Original image", "Adjusted Image", "CLAHE"]

for i in range(len(output)):
    plt.subplot(1, 3, i + 1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
