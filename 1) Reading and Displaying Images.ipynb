```python
import cv2
from matplotlib import pyplot as plt
```

## 1) Reading image using cv2

```python
imgpath = "C:/Users/User/OpenCv/misc/4.2.07.tiff"
img = cv2.imread(imgpath)
print(type(img))
print(img.dtype)
print(img.shape)
img
```

## 2) Displaying image using cv2

```python
cv2.imshow('shimla_mirch', img)
cv2.waitKey(0)
cv2.destroyAllWindows() # important part! (Keyboard intterupt)
```

```python
cv2.namedWindow('shimla_mirch', cv2.WINDOW_AUTOSIZE)
cv2.imshow('shimla_mirch', img)
cv2.waitKey(0)
cv2.destroyWindow('shimla_mirch') 
```

## 3) Displaying image using matplotlib

```python
plt.imshow(img)
plt.show()
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## 4) Transfer Image from one directory to another directory

```python
outputImg = 'C:/Users/User/OpenCv/output/4.2.07.tiff'
img = cv2.imread(imgpath, 0) ## read in grayscale mode
cv2.imwrite(outputImg, img)
```
