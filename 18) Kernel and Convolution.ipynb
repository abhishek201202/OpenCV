```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
imgpath = "C:/Users/User/OpenCv/misc/5.1.12.tiff"
img = cv2.imread(imgpath)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## 1) Identity Convolutional Matrix

```python
k = np.array(([0,0,0], [0,1,0], [0,0,0]), np.float32)
print(k)
output = cv2.filter2D(img, -1, k)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(output)
plt.title("Filtered Image")

plt.show()
```

## 2) Edge Detection

```python
k = np.array(([1,0,-1], [0,0,0], [-1,0,1]), np.float32)
print(k)
output = cv2.filter2D(img, -1, k)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(output)
plt.title("Filtered Image")

plt.show()
```

## 3) Box Blur

```python
k = np.array(([1,1,1], [1,1,1], [1,1,1]), np.float32) / 9
print(k)
output = cv2.filter2D(img, -1, k)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(output)
plt.title("Filtered Image")

plt.show()
```
