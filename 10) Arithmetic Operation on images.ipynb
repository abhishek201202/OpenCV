```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "C:/Users/User/OpenCv/misc/"
img1path = path + "4.2.06.tiff"
img2path = path + "lena_color_512.tif"

img1 = cv2.imread(img1path, 1)
img2 = cv2.imread(img2path, 1)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = img1 + img2
img4 = img1 - img2
img5 = img2 - img1
img6 = img1 + 50
img7 = img1 * img2
img8 = img1 / img2
img9 = img1 * 2
img10 = img1 / 2
```

```python
titles = ["Scenery", "lena", "addition", "img1 - img2", "img2 - img1", "img1 + 50", "mul", "div", "img1 * 2", "img1 / 2"]
images = [img1, img2, img3, img4, img5, img6, img7, img8, img9, img10]
for i in range(len(images)):
    plt.subplot(1, len(images), i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
